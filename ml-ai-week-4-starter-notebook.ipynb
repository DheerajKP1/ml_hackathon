```python
train_df = pd.read_csv("/kaggle/input/practice-hackathon/Train.csv")
```

```python
## Label encoding column A of our dataset
from sklearn import preprocessing
  
# label_encoder object knows 
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
train_df['A']= label_encoder.fit_transform(train_df['A'])
  
train_df['A'].unique()
```

```python
X = train_df.drop(['Target'], axis=1)
y = train_df['Target']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.1)
```

```python
print(y_train.value_counts())
```

```python
lr=LogisticRegression(solver='liblinear')
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_val.size,y_pred.size
```

```python
Y_val = y_val.copy()
Y_pred = y_pred.copy()
```

```python
from tabulate import tabulate
import numpy as np


# Convert the arrays to a list of lists
rows = np.vstack((Y_val,Y_pred))
rows
columns = list(zip(*rows))

# Print the columns
for column in columns:
    print(column)
```

```python
accuracy_score(y_val, y_pred)
```

```python
error= mean_squared_error(y_val, y_pred)
print(error)
```
